### 📊 Dataset Overview

This project uses multiple tables from the inventory database. Each table captures a different part of the business process such as purchasing, pricing, sales, and vendor invoices.

#### 1. Purchases Table

The **purchases** table contains actual purchase transaction data. It includes:

* Date of purchase
* Product brands purchased by vendors
* Quantity purchased
* Total amount paid (in dollars)

#### 2. Purchase Prices Table

The **purchase_prices** table contains product-level pricing information. It provides:

* Actual product price
* Purchase price offered by vendors

The combination of **Vendor + Brand** is unique in this table.

#### 3. Vendor Invoice Table

The **vendor_invoice** table aggregates information from the purchases table. It summarizes:

* Total quantity purchased
* Total purchase dollar amount
* Freight cost associated with the purchase

Uniqueness in this table is maintained using **Vendor + PO Number**.

#### 4. Sales Table

The **sales** table records actual sales transactions. It includes:

* Products sold
* Quantity sold
* Selling price
* Revenue generated from sales

---

### 📈 Data Requirement for Analysis

Since the required data is distributed across multiple tables, we need to create a **summary dataset** that combines information from different sources.

The summary table should contain:

* Purchase transactions made by vendors
* Sales transaction data
* Freight costs for each vendor
* Actual product prices from vendors

```python
# Freight cost — har vendor ne kitna freight pay kiya
# Ye baad mein main summary query mein join hoga
freight_summary = pd.read_sql_query("""
    SELECT
        VendorNumber,
        SUM(Freight) AS FreightCost
    FROM vendor_invoice
    GROUP BY VendorNumber
""", conn)
freight_summary
```

```python
# Purchase summary — vendor + brand level par total purchase quantity aur spend
# purchase_prices se join karke actual selling price aur volume bhi laaya
pd.read_sql_query("""
    SELECT
        p.VendorNumber,
        p.VendorName,
        p.Brand,
        p.PurchasePrice,
        pp.Volume,
        pp.Price AS actual_price,
        SUM(p.Quantity) AS total_quantity,
        SUM(p.Dollars) AS TotalPurchaseDollars
    FROM purchases p
    JOIN purchase_prices pp
        ON p.Brand = pp.Brand
    WHERE p.PurchasePrice > 0
    GROUP BY p.VendorNumber, p.VendorName, p.Brand
    ORDER BY TotalPurchaseDollars
""", conn)
```

```python
# Sales summary — vendor + brand level par total sales
sales_summary = pd.read_sql_query("""
    SELECT
        VendorNo,
        Brand,
        SUM(SalesDollars)  AS TotalSalesDollars,
        SUM(SalesPrice)    AS TotalSalesPrice,
        SUM(SalesQuantity) AS TotalSalesQuantity,
        SUM(ExciseTax)     AS TotalExciseTax
    FROM sales
    GROUP BY VendorNo, Brand
    ORDER BY TotalSalesDollars
""", conn)
sales_summary
```

```python
# ── MAIN QUERY — Vendor Sales Summary ───────────────────────────────────────
# Ye heavy query hai jo 3 CTEs use karti hai:
#   FreightSummary  → har vendor ka total freight cost
#   PurchaseSummary → har vendor+brand ka purchase data
#   SalesSummary    → har vendor+brand ka sales data
# Teeno ko JOIN karke ek final summary table banate hain
# Query time bhi measure kiya taaki performance track ho

import time

start = time.time()

vendor_sales_summary = pd.read_sql_query("""
    WITH FreightSummary AS (
        SELECT
            VendorNumber,
            SUM(Freight) AS FreightCost
        FROM vendor_invoice
        GROUP BY VendorNumber
    ),

    PurchaseSummary AS (
        SELECT
            p.VendorNumber,
            p.VendorName,
            p.Brand,
            p.Description,
            p.PurchasePrice,
            pp.Price AS ActualPrice,
            pp.Volume,
            SUM(p.Quantity) AS TotalPurchaseQuantity,
            SUM(p.Dollars)  AS TotalPurchaseDollars
        FROM purchases p
        JOIN purchase_prices pp
            ON p.Brand = pp.Brand
        WHERE p.PurchasePrice > 0
        GROUP BY p.VendorNumber, p.VendorName, p.Brand, p.Description, p.PurchasePrice, pp.Price, pp.Volume
    ),

    SalesSummary AS (
        SELECT
            VendorNo,
            Brand,
            SUM(SalesQuantity) AS TotalSalesQuantity,
            SUM(SalesDollars)  AS TotalSalesDollars,
            SUM(SalesPrice)    AS TotalSalesPrice,
            SUM(ExciseTax)     AS TotalExciseTax
        FROM sales
        GROUP BY VendorNo, Brand
    )

    SELECT
        ps.VendorNumber,
        ps.VendorName,
        ps.Brand,
        ps.Description,
        ps.PurchasePrice,
        ps.ActualPrice,
        ps.Volume,
        ps.TotalPurchaseQuantity,
        ps.TotalPurchaseDollars,
        ss.TotalSalesQuantity,
        ss.TotalSalesDollars,
        ss.TotalSalesPrice,
        ss.TotalExciseTax,
        fs.FreightCost
    FROM PurchaseSummary ps
    LEFT JOIN SalesSummary ss
        ON ps.VendorNumber = ss.VendorNo
        AND ps.Brand = ss.Brand
    LEFT JOIN FreightSummary fs
        ON ps.VendorNumber = fs.VendorNumber
    ORDER BY ps.TotalPurchaseDollars DESC
""", conn)

end = time.time()
display(vendor_sales_summary.head())
print("Query Time:", end - start)
```

### Vendor Sales & Purchase Summary

This query generates a **vendor-wise sales and purchase summary**, which is valuable for business analysis and reporting.

### Performance Optimization

* The query involves **heavy joins and aggregations** on large datasets such as *sales* and *purchases*.
* Storing **pre-aggregated results** helps avoid repeated expensive computations.
* It enables easier analysis of **sales, purchases, and pricing** across different vendors and brands.
* Pre-computed data improves performance for **dashboards and reporting systems**.
* Instead of running expensive queries repeatedly, dashboards can **quickly retrieve data from the `vendor_sales_summary` table**.

```python
# Missing values check karo
# Sales columns mein NULL hoga kyunki LEFT JOIN use kiya —
# kuch products purchase hue honge lekin beke nahi
vendor_sales_summary.isnull().sum()
```

```python
# Volume column mein 'Unknown' string aur NaN values hain — inhe hatao
# fillna commented out hai kyunki pehle Volume fix karna zaroori tha
vendor_sales_summary['Volume'].unique()
# vendor_sales_summary = vendor_sales_summary.fillna(0, inplace=True)  # pehle ye nahi — pehle Volume clean karo
vendor_sales_summary = vendor_sales_summary.dropna(subset=['Volume'])
vendor_sales_summary = vendor_sales_summary[vendor_sales_summary['Volume'] != 'Unknown']
```

```python
# Null check dobara — Volume clean hone ke baad
vendor_sales_summary['Description'].unique()
vendor_sales_summary.isnull().sum()
```

```python
# Volume ko string se float mein convert karo — calculations ke liye zaroori hai
vendor_sales_summary['Volume'] = vendor_sales_summary['Volume'].astype('float64')
```

```python
# Baaki NaN (sales columns) ko 0 se fill karo
# Matlab: jinhe purchase kiya lekin bika nahi unhe 0 maano
vendor_sales_summary.fillna(0, inplace=True)
```

```python
# VendorName ke aage-peeche extra spaces hatao
vendor_sales_summary['VendorName'] = vendor_sales_summary['VendorName'].str.strip()
```

```python
# Gross Profit = Total Sales Revenue - Total Purchase Cost
# Negative aaya toh matlab product bika nahi ya loss mein bika
vendor_sales_summary['GrossProfit'] = (
    vendor_sales_summary['TotalSalesDollars'] - vendor_sales_summary['TotalPurchaseDollars']
)
```

```python
# Gross Profit ka minimum dekho — negative values hain kya
vendor_sales_summary['GrossProfit'].min()
```

```python
# Profit Margin (%) = GrossProfit / TotalSalesDollars * 100
# Kitne percent revenue actually profit mein gaya
vendor_sales_summary['ProfitMargin'] = (
    vendor_sales_summary['GrossProfit'] / vendor_sales_summary['TotalSalesDollars'] * 100
)
```

```python
# Stock Turnover = Units Sold / Units Purchased
# 1.0 = sab bik gaya, <1 = kuch bacha, >1 = purana stock bhi bika
vendor_sales_summary['StockTurnover'] = (
    vendor_sales_summary['TotalSalesQuantity'] / vendor_sales_summary['TotalPurchaseQuantity']
)
```

```python
# Sales to Purchase Ratio = Revenue / Cost
# Har 1 dollar purchase par kitna revenue aaya
vendor_sales_summary['SalesToPurchaseRatio'] = (
    vendor_sales_summary['TotalSalesDollars'] / vendor_sales_summary['TotalPurchaseDollars']
)
```

```python
# Final null check — sab 0 hone chahiye ab
vendor_sales_summary.isnull().sum()
```

```python
cursor = conn.cursor()
cursor.execute("""
CREATE TABLE vendor_sales_summary (
    VendorNumber INT,
    VendorName VARCHAR(100),
    Brand INT,
    Description VARCHAR(100),

    PurchasePrice DECIMAL(10,2),
    ActualPrice DECIMAL(10,2),
    Volume FLOAT,

    TotalPurchaseQuantity INT,
    TotalPurchaseDollars DECIMAL(15,2),

    TotalSalesQuantity INT,
    TotalSalesDollars DECIMAL(15,2),
    TotalSalesPrice DECIMAL(15,2),
    TotalExciseTax DECIMAL(15,2),

    FreightCost DECIMAL(15,2),

    GrossProfit DECIMAL(15,2),
    ProfitMargin DECIMAL(15,2),
    StockTurnover DECIMAL(15,2),
    SalesToPurchaseRatio DECIMAL(15,2),

    PRIMARY KEY (VendorNumber, Brand)
);
""")

vendor_sales_summary.to_sql('vendor_sales_summary', conn, if_exists='replace', index=False)
```
